```python
import cv2
from facenet_pytorch import MTCNN, InceptionResnetV1
from PIL import Image
import torch

# Function to extract frames from a video
def extract_frames(video_path, frame_skip=10):
    cap = cv2.VideoCapture(video_path)
    frames = []
    idx = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if idx % frame_skip == 0:
            frames.append(frame)
        idx += 1
    cap.release()
    return frames

# Determine the device to use (GPU if available, otherwise CPU)
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Initialize MTCNN for face detection and InceptionResnetV1 for face embeddings
mtcnn = MTCNN(image_size=160, margin=20, device=device)
resnet = InceptionResnetV1(pretrained='vggface2').eval().to(device)

# Function to get face embeddings from extracted frames
def get_face_embeddings(frames):
    embeddings = []
    for frame in frames:
        img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        face = mtcnn(img)
        if face is not None:
            face = face.unsqueeze(0).to(device)
            emb = resnet(face).detach().cpu().numpy()
            embeddings.append(emb)
    if embeddings:
        return np.vstack(embeddings)  # shape [num_frames, 512]
    else:
        return None
```

```python
import librosa, numpy as np, os
from moviepy.editor import VideoFileClip

def extract_audio_features(video_path, sr=16000, n_mels=128):
    #Extracts spectrogram features from the audio track of a video
    try:
        clip = VideoFileClip(video_path)
        audio = clip.audio
        if audio is None:
            return None
        audio_path = "temp_audio.wav"
        audio.write_audiofile(audio_path, fps=sr, verbose=False, logger=None)

        # Load and compute mel spectrogram
        y, sr = librosa.load(audio_path, sr=sr)
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels)
        mel_db = librosa.power_to_db(mel, ref=np.max)
        return mel_db

    except Exception as e:
        print(f"Error processing {video_path}: {e}")
        return None
```

```python
class FakeAVCelebDataset(Dataset):
    def __init__(self, meta_df):
        """
        meta_df: DataFrame with columns ['Video_emb', 'Audio_emb', 'type_num']
        """
        self.meta_df = meta_df.reset_index(drop=True)

    def __len__(self):
        return len(self.meta_df)

    def __getitem__(self, idx):
        # Load embeddings
        video_path = self.meta_df.loc[idx, 'Video_emb']
        audio_path = self.meta_df.loc[idx, 'Audio_emb']

        video_emb = torch.tensor(np.load(video_path), dtype=torch.float32)  # convert video to tensors
        audio_emb = torch.tensor(np.load(audio_path).T, dtype=torch.float32)  # convert audio to tensors

        # Label
        label = torch.tensor(self.meta_df.loc[idx, 'type_num'], dtype=torch.long)

        return video_emb, audio_emb, label
```

```python
from torch.nn.utils.rnn import pad_sequence

def collate_fn(batch):
    videos, audios, labels = zip(*batch)

    # Pad sequences
    videos_padded = pad_sequence(videos, batch_first=True)
    audios_padded = pad_sequence(audios, batch_first=True)

    # Build masks directly using list comprehension
    video_masks = torch.stack([torch.arange(videos_padded.size(1)) < v.size(0) for v in videos])
    audio_masks = torch.stack([torch.arange(audios_padded.size(1)) < a.size(0) for a in audios])

    labels = torch.tensor(labels)

    return videos_padded, video_masks, audios_padded, audio_masks, labels
```

```python
import pandas as pd

df = meta_emb.copy()

#  Keep all of 2
subset_1 = df[df['type_num'].isin([1])]

#  For 1 and 3, take 100 per genderâ€“race combination
samples = []
for t in [0, 2]:
    class_df = df[df['type_num'] == t]
    grouped = class_df.groupby(['gender', 'race'])

    selected = []
    for (gender, race), group in grouped:
        # Take up to 100 deterministically (no random sampling)
        take_n = min(len(group), 200)
        selected.append(group.head(take_n))

    selected_df = pd.concat(selected, ignore_index=True)
    samples.append(selected_df)

# Combine everything
final_balanced_df = pd.concat([subset_1] + samples, ignore_index=True)

# Check distribution
print("Final sample sizes per class:")
print(final_balanced_df['type_num'].value_counts())

print("\nDetailed group breakdown:")
print(final_balanced_df.groupby(['type_num', 'gender', 'race']).size())
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class CrossAttention(nn.Module):
    """
    Cross-Attention module for fusing video and audio sequences.
    """
    def __init__(self, embed_dim, num_heads=4):
        super(CrossAttention, self).__init__()
        self.attn = nn.MultiheadAttention(
            embed_dim=embed_dim,
            num_heads=num_heads,
            batch_first=True
        )

    def forward(self, query, key, value, key_padding_mask=None):
        attn_out, attn_weights = self.attn(
            query=query,
            key=key,
            value=value,
            key_padding_mask=key_padding_mask
        )
        return attn_out, attn_weights


class MultiModalFakeDetector(nn.Module):
    def __init__(self, video_dim=512, audio_dim=128, hidden_dim=256, num_classes=3, num_heads=4):
        """
        num_classes set to 3 â†’ classes: 0,1,2 (remapped)
        """
        super(MultiModalFakeDetector, self).__init__()

        # Project embeddings to same hidden_dim
        self.video_proj = nn.Linear(video_dim, hidden_dim)
        self.audio_proj = nn.Linear(audio_dim, hidden_dim)

        # Cross-attention module
        self.cross_attn = CrossAttention(embed_dim=hidden_dim, num_heads=num_heads)

        # Classification head (3 classes)
        self.class_head = nn.Linear(hidden_dim, num_classes)

    def forward(self, video_emb, video_mask, audio_emb, audio_mask):

        # Project embeddings
        video_proj = self.video_proj(video_emb)
        audio_proj = self.audio_proj(audio_emb)

        # Cross-attention: video attends to audio
        attn_out, attn_weights = self.cross_attn(
            query=video_proj,
            key=audio_proj,
            value=audio_proj,
            key_padding_mask=~audio_mask
        )

        # Fuse attended features
        fused = video_proj + attn_out

        # Pool sequence â†’ fixed-size representation
        fused_pool = fused.permute(0, 2, 1)              # (B, hidden_dim, L)
        fused_pool = F.adaptive_avg_pool1d(fused_pool, 1).squeeze(-1)

        # Multiclass output (3 logits)
        class_out = self.class_head(fused_pool)

        return class_out, attn_weights
```

```python
# Model, device, optimizer
import torch
import torch.nn as nn
import torch.optim as optim

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Parameters
video_dim = 512
audio_dim = 128
hidden_dim = 256
num_classes = 3
num_heads = 4

model = MultiModalFakeDetector(
    video_dim=video_dim,
    audio_dim=audio_dim,
    hidden_dim=hidden_dim,
    num_classes=num_classes,
    num_heads=num_heads
)

model = model.to(device)

optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
from sklearn.utils import resample
from tqdm import tqdm

# Split data
train_df, test_df = train_test_split(
    final_balanced_df,
    test_size=0.2,
    random_state=42,
    stratify=final_balanced_df['type_num']
)

# Balance data
class_0 = train_df[train_df['type_num'] == 0]
class_1 = train_df[train_df['type_num'] == 1]
class_2 = train_df[train_df['type_num'] == 2]

target_size = max(len(class_0), len(class_1), len(class_2))

class_0_up = resample(class_0, replace=True, n_samples=target_size, random_state=42)
class_1_up = resample(class_1, replace=True, n_samples=target_size, random_state=42)
class_2_up = resample(class_2, replace=True, n_samples=target_size, random_state=42)

train_df_balanced = pd.concat([class_0_up, class_1_up, class_2_up]) \
    .sample(frac=1, random_state=42).reset_index(drop=True)

print("Balanced train class counts:")
print(train_df_balanced['type_num'].value_counts())

# Data Loaders
train_dataset = FakeAVCelebDataset(train_df_balanced)
test_dataset = FakeAVCelebDataset(test_df)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, collate_fn=collate_fn)
```

```python
class_weights = torch.tensor([2.05, 2.15, 1.35], dtype=torch.float32).to(device)
criterion = nn.CrossEntropyLoss(weight=class_weights)

# Training
for epoch in range(20):
    model.train()
    total_loss = 0

    loop = tqdm(enumerate(train_loader), total=len(train_loader), desc=f"Epoch {epoch+1}/20")

    for i, (videos, video_masks, audios, audio_masks, labels) in loop:
        videos, video_masks = videos.to(device), video_masks.to(device)
        audios, audio_masks = audios.to(device), audio_masks.to(device)
        labels = labels.to(device)   # already 0/1/2 from Dataset

        optimizer.zero_grad()

        logits, attn = model(videos, video_masks, audios, audio_masks)
        loss = criterion(logits, labels)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        loop.set_postfix(loss=total_loss/(i+1))

    print(f"Epoch {epoch+1} Avg Loss: {total_loss/len(train_loader):.4f}")
```

```python
import torch
import torch.nn as nn
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.eval()  # Set model to evaluation

all_class_labels = []
all_class_preds = []
all_class_probs = []

with torch.no_grad():
    for videos, video_masks, audios, audio_masks, labels in test_loader:  # or val_loader
        videos = videos.to(device)
        video_masks = video_masks.to(device)
        audios = audios.to(device)
        audio_masks = audio_masks.to(device)
        labels = labels.to(device)

        # Forward pass
        class_logits, attn_weights = model(videos, video_masks, audios, audio_masks)

        # Multiclass predictions
        class_probs = torch.softmax(class_logits, dim=1)
        class_preds = torch.argmax(class_logits, dim=1)

        all_class_labels.extend(labels.cpu().numpy())
        all_class_preds.extend(class_preds.cpu().numpy())
        all_class_probs.extend(class_probs.cpu().numpy())

class_acc = accuracy_score(all_class_labels, all_class_preds)
class_cm = confusion_matrix(all_class_labels, all_class_preds)

print(" Multiclass Metrics ")
print(f"Accuracy: {class_acc:.4f}")
print("Confusion Matrix:")
print(class_cm)
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import precision_recall_curve, average_precision_score

def plot_multiclass_pr(all_labels, all_probs, class_names=None):
    num_classes = all_probs.shape[1]
    y_true_bin = label_binarize(all_labels, classes=list(range(num_classes)))

    plt.figure(figsize=(8, 6))

    for i in range(num_classes):
        precision, recall, _ = precision_recall_curve(y_true_bin[:, i], all_probs[:, i])
        avg_prec = average_precision_score(y_true_bin[:, i], all_probs[:, i])
        label = f"Class {i}" if class_names is None else class_names[i]
        plt.plot(recall, precision, lw=2, label=f"{label} (AP={avg_prec:.3f})")

    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title("Per-Class Precision-Recall Curves")
    plt.legend(loc="lower left")
    plt.grid(True)
    plt.show()

plot_multiclass_pr(all_class_labels, all_class_probs, class_names=['FakeVideo-RealAudio', 'RealVideo-FakeAudio', 'FakeVideo-FakeAudio'])
```
